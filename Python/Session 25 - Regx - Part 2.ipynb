```python
import re
```

```python
r = re.compile("^[a-z0-9]+([\._][a-z0-9]+)*@[a-z0-9]+(\.[a-z]{2,3}){1,2}$")
m = re.search(r,f)
if m:
    print(m.group())
else:
    print("Invlid email")
```

```python
r = re.compile("\.")
s = ".abc"
m = re.search(r,s)
if m:
    print(m.group())
else:
    print("Invalid")
```

```python
a = "http://www.xyz-a.com"
r = re.compile("^https?://www\.\w+(-\w+)*(\.[a-z]{2,3}){1,2}$")
m = re.search(r,a)
if m:
    print(m.group())
else:
    print("Invalid URL")
```
